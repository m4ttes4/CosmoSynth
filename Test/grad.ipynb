```python
import autograd.numpy as np
from autograd import grad


# Definizione della funzione gaussiana 2D
def gaussian_2d(x, y, mean_x, mean_y, stddev_x, stddev_y):
    normalization = 1 / (2 * np.pi * stddev_x * stddev_y)
    exponent = -(
        (x - mean_x) ** 2 / (2 * stddev_x**2) + (y - mean_y) ** 2 / (2 * stddev_y**2)
    )
    return normalization * np.exp(exponent)


# Definizione della funzione di perdita
def loss_function(params):
    mean_x, mean_y, stddev_x, stddev_y = params
    return -gaussian_2d(x, y, mean_x, mean_y, stddev_x, stddev_y)


# Definizione dei parametri iniziali
params = np.array(
    [0.0, 0.0, 2.0, 2.0]
)  # Inizializzazione di mean_x, mean_y, stddev_x, stddev_y
x = 1.0
y = 2.0

# Impostazioni dell'algoritmo di gradient descent
learning_rate = 0.01
num_iterations = 10000
tolerance = 1e-6  # Tolleranza per la condizione di uscita

# Calcolo del gradiente della funzione di perdita rispetto ai parametri
gradient_loss_function = grad(loss_function)

# Iterazioni di gradient descent
previous_loss = loss_function(params)
for i in range(num_iterations):
    grad_values = gradient_loss_function(params)
    params -= learning_rate * grad_values  # Aggiornamento dei parametri

    current_loss = loss_function(params)
    loss_change = np.abs(current_loss - previous_loss)

    if i % 10 == 0:  # Stampa il valore ogni 10 iterazioni
        print(
            f"Iteration {i}, params: {params}, Gradients: {grad_values}, Loss: {current_loss}"
        )

    if loss_change < tolerance:
        print(f"Convergence reached at iteration {i}. Loss change: {loss_change}")
        break

    previous_loss = current_loss

print(
    f"Optimized parameters: mean_x={params[0]}, mean_y={params[1]}, stddev_x={params[2]}, stddev_y={params[3]}"
)
```

```python
from functools import wraps


def kwargs_wrapper(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        if len(args) == 1 and isinstance(args[0], dict):
            # Se viene passato un solo argomento ed è un dizionario, usalo come kwargs
            kwargs.update(args[0])
        elif len(args) > 0:
            raise TypeError(
                "Se vengono passati argomenti posizionali, devono essere solo kwargs."
            )
        return func(**kwargs)

    return wrapper


# Esempio di utilizzo del wrapper
@kwargs_wrapper
def func(a=None, b=None, c=None):
    print(f"a={a}, b={b}, c={c}")


# Chiamata con kwargs
func(a=0, b=1)

# Chiamata con un dizionario
params = {"a": 0, "b": 1}
func(params)

# Chiamata con kwargs e un dizionario
func(params, c=2)
```
